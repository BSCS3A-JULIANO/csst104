1. **creating** **datasets**

```python
import pandas as pd

data = {
    'student_id': range(1,21),
    'gender': ['Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female','Male','Female'],
    'test_data': pd.date_range(start='2024-02-21',periods=20),
    'study_hours': [2,3,2,2,3,2,1,1,2,2,3,3,2,5,4,6,3,5,1,6],
    'test_score': [88,84,91,75,76,84,82,94,91,90,80,81,80,83,75,77,88,79,74,75,],
    'attendance': ['present','absent','present','present','present','absent','present','present','present','absent','present','present','present','absent','present','present','present','absent','present','present'],


}

df = pd.DataFrame(data)

df.to_csv('student_grade.csv',index=False)
```

2. **data loading and initial examination**

```python
#Load dataSet
df = pd.read_csv('student_grade.csv')

print(df.head(13))

df.isnull().sum()
```

**3. exploratory data analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score','study_hours']].describe()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins=15,kde=True)
plt.title('distribution of test score')

plt.subplot(1,2,2)
sne.histplot(df['study_hours'],bins=15,kde=True)
plt.title('distribution of study hours')

plt.figure(figsize=(8,6))
sne.scatterplot(data=df,x='study_hours',y='test_score')
plt.title('Scatter plot of study hours vs. test scores')
plt.grid(True)
```

**4. Data Visualization**

```python
plt.figure(figsize=(6,6))
sne.countplot(data=df,x='gender')
plt.title('gender distribution')

plt.figure(figsize=(20,6))
sne.lineplot(data=df,x='test_data',y='test_score')
plt.title('trend in test score over time')

correlation_matrix = df[['study_hours','test_score','attendance']].corr()

plt.figure (figsize = (8,6))
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title ('correlation heatmap')
```
