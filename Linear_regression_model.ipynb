#**1. import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#**2. importing csv**

```python
data = pd.read_csv('Lesson_4_student_grades_dataset 3.csv')
data
```

#**3. visualizing the data**

```python
plt.scatter(data['Previous_Exam_Scores'],data['Grades'],color='gray')
plt.xlabel('Independent variable (X)')
plt.ylabel('Dependent variable (Y)')
plt.title('Scatter plot of data')
plt.show()
```

#**4. Preparing data for modeling**

```python
X = data[['Previous_Exam_Scores']] #independent variable
Y = data[['Grades']] #dependent variable
```

#**5. Fitting Linear regression model**

```python
model = LinearRegression()
model.fit(X,Y)
```

#**6. getting model coefficient**

```python
slope = model.coef_[0]

intercept = model.intercept_

print('slope :',slope)
print('intercept :',intercept)
```

#**7. visualizing linear regression**

```python
plt.scatter(X,Y,color='gray')
plt.plot(X,model.predict(X),color='red')
plt.xlabel('Independent variable (X)')
plt.ylabel('Dependent variable (Y)')
plt.title('Scatter plot of data')
plt.show()
```

#**8. Making predictions**

```python
new_X = [[np.random.randint(100,200)]]

predicted_Y = model.predict(new_X)

print('Value of X:', new_X)
print('Predicted Value of Y:', predicted_Y)
```
